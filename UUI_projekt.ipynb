```python
from itertools import product

class Room:
    def __init__(self, name):
        self.name = name

class Teacher:
    def __init__(self, name, subjects):
        self.name = name
        self.subjects = subjects

class Subject:
    def __init__(self, name):
        self.name = name

class ClassSession:
    def __init__(self, subject, teacher, room, time_slot, group):
        self.subject = subject
        self.teacher = teacher
        self.room = room
        self.time_slot = time_slot
        self.group = group

class Schedule:
    def __init__(self, teachers, rooms, subjects, time_slots, groups):
        self.teachers = teachers
        self.rooms = rooms
        self.subjects = subjects
        self.time_slots = time_slots
        self.groups = groups
        self.classes = []




    def add_class(self, class_session):
        self.classes.append(class_session)

    def is_valid(self):
        for i in range(len(self.classes)):
            for j in range(i + 1, len(self.classes)):
                class1 = self.classes[i]
                class2 = self.classes[j]


                if class1.teacher == class2.teacher and class1.time_slot == class2.time_slot:
                    return False
                if class1.room == class2.room and class1.time_slot == class2.time_slot:
                    return False


                if class1.group == class2.group and class1.time_slot == class2.time_slot:
                    return False

        return True

    def schedule_classes_dfs(self, class_index=0):

        if class_index == len(self.subjects) * len(self.groups):
            return True


        subject_index = class_index // len(self.groups)
        group_index = class_index % len(self.groups)
        subject = self.subjects[subject_index]
        group = self.groups[group_index]

        for teacher, room, time_slot in product(self.teachers, self.rooms, self.time_slots):
            if subject in teacher.subjects:
                class_session = ClassSession(subject, teacher, room, time_slot, group)
                self.add_class(class_session)

                if self.is_valid():
                    if self.schedule_classes_dfs(class_index + 1):
                        return True

                self.classes.pop()

        return False


def print_schedule(schedule):

    def parse_time_slot(time_slot):
        days = ["Ponedjeljak", "Utorak", "Srijeda", "Cetvrtak", "Petak"]
        day, time = time_slot.split(" ", 1)
        hour = int(time.split(":")[0])
        day_index = days.index(day)
        return day_index, hour


    for group in schedule.groups:
        print(f"Raspored za grupu {group}:\n")
        group_classes = [c for c in schedule.classes if c.group == group]
        sorted_classes = sorted(group_classes, key=lambda x: parse_time_slot(x.time_slot))


        for class_session in sorted_classes:
            print(f"Time Slot: {class_session.time_slot}, "
                  f"Subject: {class_session.subject.name}, "
                  f"Teacher: {class_session.teacher.name}, "
                  f"Room: {class_session.room.name}")
        print("\n" * 2)



matematika = Subject("Matematika")
biologija = Subject("Biologija")
hrvatski = Subject("Hrvatski")
glazbeni = Subject("Glazbeni")
kemija = Subject("Kemija")
latinski = Subject("Latinski")
fizika = Subject("Fizika")
njemacki = Subject("Njemacki")
psihologija = Subject("Psihologija")
sociologija = Subject("Sociologija")
logika = Subject("Logika")

d1 = Room("D1")
d2 = Room("D2")
d3 = Room("D3")
d4 = Room("D4")
d5 = Room("D5")


teacher1 = Teacher("Antonija Maric", [matematika, biologija])
teacher2 = Teacher("Lidija Matic", [hrvatski])
teacher3 = Teacher("Mario Matosic", [hrvatski, latinski])
teacher4 = Teacher("Lea Kalinic", [kemija])
teacher5 = Teacher("Tena Sakic", [fizika])
teacher6 = Teacher("Marko Radic", [njemacki])
teacher7 = Teacher("Toni Saric", [psihologija, logika])
teacher8 = Teacher("Mario Vlasic", [sociologija])
teacher9 = Teacher("Maja Krovinovic", [logika])


teachers = [teacher1, teacher2, teacher3, teacher4, teacher5, teacher6, teacher7, teacher8, teacher9]
rooms = [d1, d2, d3, d4, d5]
subjects = [matematika, biologija, hrvatski, latinski, kemija, fizika, njemacki, psihologija, sociologija, logika]



groups = ["4a", "4b", "4c", "4d"]

time_slots = ["Ponedjeljak 8:00 - 10:00 ", "Ponedjeljak 10:00 - 12:00", "Ponedjeljak 12:00 - 14:00", "Utorak 8:00 - 9:00", "Utorak 9:30 - 11:00", "Utorak 11:15 - 12:00", "Srijeda 8:00 - 10:00", "Srijeda 10:00 - 11:30", "Srijeda 12:00 - 13:30", "Cetvrtak 8:00 - 10:00", "Cetvrtak 10:00 - 12:00", "Cetvrtak 12:00 - 14:00", "Petak 8:00 - 9:00", "Petak 9:00 - 11:00", "Petak 11:00 - 13:00"]
schedule = Schedule(teachers, rooms, subjects, time_slots, groups)


if schedule.schedule_classes_dfs():
    print_schedule(schedule)
else:
    print("Raspored nije pronadjen.")
```
